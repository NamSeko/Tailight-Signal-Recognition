```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
save_model_path = './models'

A = np.load(os.path.join(save_model_path, 'training_loss.npy'))
B = np.load(os.path.join(save_model_path, 'training_score.npy'))
C = np.load(os.path.join(save_model_path, 'test_loss.npy'))
D = np.load(os.path.join(save_model_path, 'test_score.npy'))
epochs = len(A)

fig = plt.figure(figsize=(20, 5))
plt.subplot(121)
plt.plot(np.arange(1, epochs + 1), A)
plt.plot(np.arange(1, epochs + 1), C)
plt.title("Losses")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc="upper left")

plt.subplot(122)
plt.plot(np.arange(1, epochs + 1), B)
plt.plot(np.arange(1, epochs + 1), D)
plt.title("Scores")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc="upper left")
plt.show()
```

```python
best_train_score = np.max(B)
print(f"Best train score: {best_train_score:.4f} at epoch {np.argmax(B) + 1}")
```

```python
best_epochs = np.where(D == np.max(D))[0]
for epoch in best_epochs:
    print('Best epoch: {}, Validation Accuracy: {:.2f}%'.format(epoch + 1, 100 * D[epoch]))
```
